## Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# X.isnull().sum()
```

## Linear Regression

```python
#Fitting the Simple Linear Regression model to the training dataset  
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions=lm.predict(X_test)
```

```python
print(lm.score(X_test, y_test))
```

## Evaluation

```python
import numpy as np   # Evaluation 
import sklearn.metrics as metrics
print('MAE: {}'.format(metrics.mean_absolute_error(y_test, predictions)))
print('MSE: {}'.format(metrics.mean_squared_error(y_test, predictions)))
print('RMSE: {}'.format(np.sqrt(metrics.mean_squared_error(y_test, predictions))))
```

```python
print(y_test.max())
print(y_test.min())
```

## BY Using RIDGE

```python
from sklearn.linear_model import Ridge
lm = Ridge()
lm.fit(X_train,y_train)
```

```python
y_pred=lm.predict(X_test)
print(y_pred)
```

```python
print(lm.score(X_test, y_test))
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np

r2=r2_score(y_test,y_pred)
print("R-squared:",r2)

rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print("RMSE:",rmse)

adjusted_r_squared = 1 - (1-r2)*(len(y)-1)/(len(y)-X.shape[1]-1)
print("Adj R-square:",adjusted_r_squared)
```

## DecisionTree Regressor

```python
# predicting using the Decision_Tree_Regressor
from sklearn.tree import DecisionTreeRegressor

model_DecisionTree=DecisionTreeRegressor(criterion="friedman_mse",random_state=10,splitter="best")
# for pruning we used these (min_samples_leaf=2,min_samples_split=5,max_depth=10)


# fit the model on the data and predict the valu
model_DecisionTree.fit(X_train,y_train)
y_pred=model_DecisionTree.predict(X_test)

# print(Y_pred)
# print(list(zip(Y_test,Y_pred)))
```

```python
print(model_DecisionTree.score(X_test, y_test))
```

## Model Evalution

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np

r2=r2_score(y_test,y_pred)
print("R-squared:",r2)

rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print("RMSE:",rmse)

adjusted_r_squared = 1 - (1-r2)*(len(y)-1)/(len(y)-X.shape[1]-1)
print("Adj R-square:",adjusted_r_squared)
```

## BY USING RANDOM FOREST

```python
# predicting using the Random_Forest_Regressor
from sklearn.ensemble import RandomForestRegressor

model_RandomForest=RandomForestRegressor(n_estimators=100,random_state=10,bootstrap=True,max_features=0.6)
# n_estimators=100 mean the no.fo bags is 100

# fit the model on the data and predict the valu
model_RandomForest.fit(X_train,y_train)
y_pred=model_RandomForest.predict(X_test)
```

```python
print(model_RandomForest.score(X_test, y_test))
```

## Model Evalution
